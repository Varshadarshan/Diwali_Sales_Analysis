```python
# Import libraries
import numpy as np    # Works on array
import pandas as pd   # data freams
import matplotlib.pyplot as plt  # These two on data visualization
%matplotlib inline
import seaborn as sns
```

```python
# import csv file
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
# df = data freams
#  encoding= 'unicode_escape') = parameter (optional)
```

```python
# Gives the number of rows and column
df.shape  
```

```python
df.head() 
# Automatically give top 5 values of table
```

```python
df.info()
# info() provides info about table about the columns, data types of records, null or not null values
```

```python
# drop irrevlevent/blank row
# inplace = True) = ensures correct implementation (drop columns) and save all the changes
df.drop(['Status' , 'unnamed1'] , axis = 1, inplace = True)
```

```python
df.info()
# Confirm whether records has been dropped or not
```

```python
pd.isnull(df).sum()
# to count null values in each column
# Here we have amount col has 12 null values
```

```python
df.shape
```

```python
df.dropna(inplace = True)   
# Drop all null values (na)
```

```python
df.shape  # Checked if numbers dropped or not
```

```python
# Wanna ensure the null vales again 
pd.isnull(df).sum()  # Amount colmun has 0 null values
```

```python
# Change data type (Amount)
# astype = a function used to chnage data types
df['Amount'] = df['Amount'].astype ('int')
```

```python
df['Amount'].dtype
```

```python
# Give columns names
df.columns
```

```python
# rename column (just for self info)
df.rename(columns= {'Cust_name':'Customer name'})
```

```python
# describe() method returns the descrption of all colunns (min, max, avg, std deviation etc)
df.describe()
```

```python
# Want to use describe function on specific columns
df[['Age', 'Orders' , 'Amount']].describe()
```

```python
# Exploratory Data Analysis


# Gender 
# Gender and its count
# coontainer = to show the actual values o male and female ie. 7832 and 3407
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Purchase = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender', y = 'Amount', data = Purchase)
```

```python
From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men 
```

```python
# Age Group

# Count the numbers of age group
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bar in ax.containers:
    ax.bar_label(bars)

# Hue = distributing the numbers of Age group on the basis of gender
# if hue is not in code then chart will be the count of age group 
```

```python
# Total amount VS Age Froup

Amount_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Age Group', y = 'Amount', data = Amount_age)
```

```python
From above graphs we can see that most of the buyers are of age group between 26-35 yrs female
```

```python
# State

# No of ordwrs from Top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})   # Set the dimension
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

```python
From above graphs we can see that most of the orders & total amount are from Uttar Pradesh, Maharashtra and Karnataka respectively
```

```python
# Martial Status

ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Amount spent VS Martial Status on the basis of Gender
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

```python
From above graphs we can see that most of the buyers are married (women) and they have high purchasing power
```

```python
# Occupation

# Count of occupation
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Amount occupation wise

sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

```python
From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector
```

```python
# Product Category

sns.set(rc={'figure.figsize':(22,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Amount from different Occupations

Amount_occupations = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = Amount_occupations, x = 'Product_Category',y= 'Amount')
```

```python
From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category
```

```python
# Products VS Orders

sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

```python
Conclusion:

Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and
Aviation are more likely to buy products from Food, Clothing and Electronics category
```
